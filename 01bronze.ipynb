```python
campaigns = spark.read.csv('/Volumes/workspace/default/accenture_interju/campaigns.csv', header=True, inferSchema=True)
```

```python
customers = spark.read.csv('/Volumes/workspace/default/accenture_interju/customers.csv', header=True, inferSchema=True)
```

```python
from pyspark.sql.functions import current_timestamp, lit

customers = customers.withColumn('ingested_at', current_timestamp())
campaigns = campaigns.withColumn('ingested_at', current_timestamp())

customers = customers.withColumn('source_file', lit('customers.csv'))
campaigns = campaigns.withColumn('source_file', lit('campaigns.csv'))

customers.write.format("delta").mode("overwrite").saveAsTable("workspace.bronze.customers_raw")
campaigns.write.format("delta").mode("overwrite").saveAsTable("workspace.bronze.campaigns_raw")
```
