```python
silver_customers = spark.table('workspace.silver.customers')
silver_campaigns = spark.table('workspace.silver.campaigns')
```

```python
from pyspark.sql.functions import count, sum, round, col

campaign_performance = silver_campaigns.groupBy('campaign_name').agg(
        count('event_id').alias('total_sent'),
        sum(col('opened_at').cast('int')).alias('total_opened'),
        sum(col('clicked').cast('int')).alias('total_clicked'),
        sum(col('converted').cast('int')).alias('total_converted'),
        round(sum('revenue'), 2).alias('total_revenue')
    )
```

```python
channel_roi = silver_campaigns.groupBy('channel').agg(
        count('event_id').alias('total_sent'),
        round(sum('revenue'), 2).alias('total_revenue'),
        round(sum('revenue') / count('event_id'), 2).alias('avg_revenue_per_send')
    )
```

```python
from pyspark.sql.functions import avg

city_segments = silver_campaigns.join(silver_customers, 'customer_id') \
    .groupBy('city') \
    .agg(
        round(avg('age'), 1).alias('avg_age'),
        round(avg(col('converted').cast('int')), 3).alias('conversion_rate'),
        round(sum('revenue'), 2).alias('total_revenue')
    )
```

```python
campaign_performance.write.format("delta").mode("overwrite").saveAsTable("workspace.gold.campaign_performance")
channel_roi.write.format("delta").mode("overwrite").saveAsTable("workspace.gold.channel_roi")
city_segments.write.format("delta").mode("overwrite").saveAsTable("workspace.gold.city_segments")
```
