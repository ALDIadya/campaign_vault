```python
from faker import Faker
fake = Faker('hu_HU')
Faker.seed(42)
```

```python
n=1000
```

```python
from unidecode import unidecode
import uuid
```

```python
contacts = {}
user_name_counter = {}
email_counter = {}

for i in range(n):
    first_name = fake.first_name()
    last_name = fake.last_name()
    user_name = f"{unidecode(first_name)[0].lower()}{unidecode(last_name).lower()}"
    email_address = f"{unidecode(first_name).lower()}.{unidecode(last_name).lower()}@{fake.free_email_domain()}"

    if email_address in email_counter:
        email_counter[email_address] += 1
        email_address = f"{email_address}{email_counter[email_address]}"
    else:
        email_counter[email_address] = 1

    if user_name in user_name_counter:
        user_name_counter[user_name] += 1
        user_name = f"{user_name}{user_name_counter[user_name]}"
    else:
        user_name_counter[user_name] = 1
        user_name = user_name

    contacts[i] = {
        'customer_id': str(uuid.uuid4()),
        'first_name': first_name,
        'last_name': last_name,
        'email': email_address,
        'phone_number': fake.phone_number(),
        'user_name': user_name,
        'age': fake.random_int(min=18, max=80)
    }

print([contact['user_name'] for contact in list(contacts.values())[:]])
```

```python
customer_ids = [contacts[i]['customer_id'] for i in range(n)]
```

```python
campaigns = {}

for i in range(n):
    campaigns[i] = {
        'customer_id': fake.random_element(customer_ids),
        'campaign_name': fake.catch_phrase(),
        'currency_code': fake.currency_code(),
        'revenue': round(fake.random_number(digits=5, fix_len=True) * fake.random.random(), 2),
        'channel': fake.random_element(['email', 'SMS', 'social', 'push']),
        'event_id': fake.bothify('CAMP-####-??'),
        'clicked': fake.boolean(chance_of_getting_true=50),
        'converted': fake.boolean(chance_of_getting_true=30)

    }

print(campaigns)
```

```python
from datetime import timedelta
from random import randint
```

```python
date_times = {}

for i in range(n):
    sent = fake.date_time_between(start_date='-2y', end_date='-1y')
    opened = sent + timedelta(days=randint(1, 365))

    date_times[i] = {
        'sent_at': sent.strftime('%Y-%m-%d %H:%M:%S'),
        'opened_at': opened.strftime('%Y-%m-%d %H:%M:%S'),
        'this_year': fake.date_time_this_year().strftime('%Y'),
        'unix_time': fake.unix_time()
    }

print([date_time['opened_at'] for date_time in list(date_times.values())[:5]])
```

```python
details = {}

for i in range(n):
    details[i]= {
        'company_name': fake.company(),
        'business_details': fake.bs(),
        'catch_phrase': fake.catch_phrase(),
        'fake_sentence': fake.sentence(),
        'product_id': fake.bothify('PROD-####-??')
    }

print(details)
```

```python
geography = {}
hu_cities = [
    ('47.4979', '19.0402', 'Budapest', 'HU', 'Europe/Budapest'),
    ('47.5316', '21.6273', 'Debrecen', 'HU', 'Europe/Budapest'),
    ('46.2530', '20.1414', 'Szeged', 'HU', 'Europe/Budapest'),
    ('47.6875', '17.6504', 'Győr', 'HU', 'Europe/Budapest'),
    ('46.0727', '18.2323', 'Pécs', 'HU', 'Europe/Budapest'),
    ('47.9025', '19.6197', 'Miskolc', 'HU', 'Europe/Budapest'),
    ('47.0833', '17.9167', 'Veszprém', 'HU', 'Europe/Budapest'),
    ('46.8964', '19.6897', 'Kecskemét', 'HU', 'Europe/Budapest'),
    ('47.7536', '18.7997', 'Esztergom', 'HU', 'Europe/Budapest'),
    ('48.0000', '20.7833', 'Eger', 'HU', 'Europe/Budapest'),
    ('46.3590', '17.7956', 'Kaposvár', 'HU', 'Europe/Budapest'),
    ('47.3495', '19.8347', 'Cegléd', 'HU', 'Europe/Budapest'),
    ('48.1117', '20.7906', 'Miskolc', 'HU', 'Europe/Budapest'),
    ('46.6833', '17.6356', 'Nagykanizsa', 'HU', 'Europe/Budapest'),
    ('47.6817', '16.5845', 'Sopron', 'HU', 'Europe/Budapest'),
]

for i in range(n):
    location = fake.random_element(hu_cities)
    geography[i] = {
        'city': location[2],
        'country': location[3],
        'latitude': location[0],
        'longitude': location[1],
        'postal_code': fake.postcode(),
        'timezone': location[4]
    }

print(geography[i])

```

```python
import pandas as pd
```

```python
contacts_df = pd.DataFrame.from_dict(contacts, orient='index')
campaigns_df = pd.DataFrame.from_dict(campaigns, orient='index')
date_times_df = pd.DataFrame.from_dict(date_times, orient='index')
details_df = pd.DataFrame.from_dict(details, orient='index')
geography_df = pd.DataFrame.from_dict(geography, orient='index')
```

```python
customers_df = pd.concat([contacts_df, geography_df], axis=1)
customers_df.to_csv('customers.csv', index=False)
print('saving is done')
```

```python
campaigns_df_full = pd.concat([campaigns_df, date_times_df, details_df], axis=1)
campaigns_df_full.to_csv('campaigns.csv', index=False)
print('saving is done')
```
